```python
import xlrd
import sys
Length = 500 #模板長度
Width = 300 #模板寬度



    

def transform(number):
    number = int(number)
    if (number == 1):
        f.write(str("\\x05\\x00\\x00\\x00\\x00\\x00\\x00\\x00\\x80\\xff?'"))
    
    elif (number >= 2 and number <= 3) :
        number = number*64
        f.write(str("\\x05\\x00\\x00\\x00\\x00\\x00\\x00\\x00"))
        f.write(str("\\"))
        f.write(hex(number)[1:])
        f.write(str("\\x00@\'"))
    elif (number >= 4 and number <= 7) :
        number = number*32
        f.write(str("\\x05\\x00\\x00\\x00\\x00\\x00\\x00\\x00"))
        f.write(str("\\"))
        f.write(hex(number)[1:])
        f.write(str("\\x01@\'"))
    elif (number >= 8 and number <= 15) :
        number = number*16
        f.write(str("\\x05\\x00\\x00\\x00\\x00\\x00\\x00\\x00"))
        f.write(str("\\"))
        f.write(hex(number)[1:])
        f.write(str("\\x02@\'"))
    elif (number >= 16 and number <= 31) :
        number = number*8
        f.write(str("\\x05\\x00\\x00\\x00\\x00\\x00\\x00\\x00"))
        f.write(str("\\"))
        f.write(hex(number)[1:])
        f.write(str("\\x03@\'"))
    elif (number >= 32 and number <= 63) :
        number = number*4
        f.write(str("\\x05\\x00\\x00\\x00\\x00\\x00\\x00\\x00"))
        f.write(str("\\"))
        f.write(hex(number)[1:])
        f.write(str("\\x04@\'"))
    elif (number >= 64 and number <= 127) :
        number = number*2
        f.write(str("\\x05\\x00\\x00\\x00\\x00\\x00\\x00\\x00"))
        f.write(str("\\"))
        f.write(hex(number)[1:])
        f.write(str("\\x05@\'"))
    elif (number >= 128 and number <= 255) :
        f.write(str("\\x05\\x00\\x00\\x00\\x00\\x00\\x00\\x00"))
        f.write(str("\\"))
        f.write(hex(number)[1:])
        f.write(str("\\x06@\'"))
    elif (number >= 256 and number <= 511) :
        reminder = number % 2 #餘數
        number = number // 2 #商數
        if(reminder == 0):
            f.write(str("\\x05\\x00\\x00\\x00\\x00\\x00\\x00\\x00"))
            f.write(str("\\"))
            f.write(hex(number)[1:])
            f.write(str("\\x07@\'"))
        else:
            f.write(str("\\x05\\x00\\x00\\x00\\x00\\x00\\x00\\x80"))
            f.write(str("\\"))
            f.write(hex(number)[1:])
            f.write(str("\\x07@\'"))
    elif (number >= 512 and number <= 1023) :
        reminder = number % 4 #餘數
        number = number // 4 #商數
        if(reminder == 0):
            f.write(str("\\x05\\x00\\x00\\x00\\x00\\x00\\x00\\x00"))
            f.write(str("\\"))
            f.write(hex(number)[1:])
            f.write(str("\\x08@\'"))
        elif(reminder == 1):
            f.write(str("\\x05\\x00\\x00\\x00\\x00\\x00\\x00@"))
            f.write(str("\\"))
            f.write(hex(number)[1:])
            f.write(str("\\x08@\'"))
        elif(reminder == 2):
            f.write(str("\\x05\\x00\\x00\\x00\\x00\\x00\\x00\\x80"))
            f.write(str("\\"))
            f.write(hex(number)[1:])
            f.write(str("\\x08@\'"))
        else:
            f.write(str("\\x05\\x00\\x00\\x00\\x00\\x00\\x00\\xC0"))
            f.write(str("\\"))
            f.write(hex(number)[1:])
            f.write(str("\\x08@\'"))

wb = xlrd.open_workbook('two_blocks.xlsx')
sh_b = wb.sheet_by_name('sheet') # 讀取base工作錶
path = 'two_blocks.txt'
f = open(path,'w')


rows = sh_b.nrows # 判斷列的個數
j = 0
i = 0
lines = 0
line = 0
while j < rows :
    col = sh_b.ncols
    line = (col - 2) // 2
    lines = lines + line
    j = j + 1

f.write(str(b'TPF0\x10TProgramHeaderV'))
f.write('\n')
f.write(str("b'2\\x00\\x0eProduct_Length"))
transform(Length)
f.write('\n')
f.write(str("b'\\rProduct_Width"))
transform(Width)
f.write('\n')
f.write(str(b'\x07Fixture\x06\x06EVP019\x00\x00TPF0\x0eTPr'))
f.write('\n')
f.write(str(b'ocedureList\x00\x08Capacit'))
f.write('\n')
f.write(str(b'y\x02\x00\x05Count\x02\x00\rComponen'))
f.write('\n')
f.write(str(b'tList\x00\x00TPF0\x0cTPattern'))
f.write('\n')
f.write(str(b'List\x00\x08Capacity\x02\x04\x05Cou'))
f.write('\n')
f.write(str(b'nt\x02\x01\rComponentList\x12T'))
f.write('\n')
f.write(str(b'DispensePatternV5\x00\x08C'))
f.write('\n')
f.write(str("b'apacity\\x02\\x04\\x05Count\\x02"))
f.write(str("\\"))
f.write(hex(lines)[1:])
f.write(str("\\x07Caption\\x06\\x04MAIN\\x0ePattern_'"))
f.write('\n')
f.write(str("b'Length"))
transform(Length)
f.write('\n')
f.write(str("b'\\rPattern_Width"))
transform(Width)
f.write('\n')
f.write(str(b'\rComponentList'))
f.write('\n')


while i < rows :
    cols = sh_b.ncols # 判斷第i列長度
    data = 0;
    while data < cols-3 :
        Bx = sh_b.cell(i, data).value 
        By = sh_b.cell(i, data+1).value 
        Ex = sh_b.cell(i, data+2).value 
        Ey = sh_b.cell(i, data+3).value 
        
        f.write(str(b'\x12TCoatInstructionV2\x00'))#線頭
        f.write('\n')
        f.write(str("b'\\x07Begin_X"))
        transform(Bx)
        f.write('\n')
        f.write(str("b'\\x07Begin_Y"))
        transform(By)
        f.write('\n')
        f.write(str("b'\\x05End_X"))
        transform(Ex)
        f.write('\n')
        f.write(str("b'\\x05End_Y"))
        transform(Ey)
        f.write('\n')
        f.write(str(b'\nSkip_Masks\x08'))#線尾
        f.write('\n')
        f.write(str(b'\x04Tool\x06\x10SC350 Applica'))
        f.write('\n')
        f.write(str(b'tor\x0fApproach_Height\x05'))
        f.write('\n')
        f.write(str(b'\x00\x00\x00\x00\x00\x00@\x9c\x0c\xc0\rDepart_He'))
        f.write('\n')
        f.write(str(b'ight\x05\x00\x00\x00\x00\x00\x00@\x9c\x0c\xc0\x0fDisp'))
        f.write('\n')
        f.write(str(b'ense_Height\x05\x00033333\xcb'))
        f.write('\n')
        f.write(str(b'\x02@\x0eDispense_Speed\x05\x00\x00'))
        f.write('\n')
        f.write(str(b'\x00\x00\x00\x00\x00\xfe\x06@\nPass_Width\x05'))
        f.write('\n')
        f.write(str(b'\x00hfffff\x98\x03@\x0bCustomBeg'))
        f.write('\n')
        f.write(str(b'in\x08\tCustomEnd\x08\x0bCusto'))
        f.write('\n')
        f.write(str(b'mSpeed\x08\x00\x00'))
        f.write('\n')
        data = data + 2
    i = i + 1

f.write(str(b'\x06Layout\x0fTRe'))
f.write('\n')
f.write(str(b'ferenceFrame\x00\x00\x00\x00\x00\x00\x00'))
f.write('\n')
f.close()
```
